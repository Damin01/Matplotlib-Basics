## Part 2 - Plotting multiple graphs on same axis

```python
fig, ax = plt.subplots()
ax.plot(x,y)
ax.plot(x,pow(x, 3))
plt.show()
```

### Label

```python
fig, ax = plt.subplots()
ax.plot(x,y, label = 'y = x')
ax.plot(x,pow(x, 3), label = 'y = $x^{2}$')
ax.legend()
plt.show()
```

## Part 3 - Scatterplot, coloring by labels

```python
fig, ax = plt.subplots()
ax.plot(x,y, label = 'y = x', c = 'red')
ax.plot(x,pow(x, 3), label = 'y = $x^{2}$', c = 'lightgray')
ax.legend()
plt.show()
```

## Part 5 - Adding and Altering Text

```python
fig, ax = plt.subplots()
ax.plot(x,y)

ax.set_xlabel('Student')
ax.set_ylabel('Banana')
ax.set_title('Student vs. Banana Relationship')

plt.show()
```

## Part 6 - Using Subplot

```python
fig, axs = plt.subplots(2, 2, figsize = (12, 8))

axs[0, 0].plot(x,y)
axs[0, 0].set_xlabel('Student')
axs[0, 0].set_ylabel('Banana')
axs[0, 0].set_title('Student vs. Banana Relationship')

axs[0, 1].plot(x,pow(y, 2))

axs[0, 1].set_xlabel('Student')
axs[0, 1].set_ylabel('Banana')
axs[0, 1].set_title('Student vs. Banana Relationship')

axs[1, 0].plot(x,pow(y, 3))

axs[1, 0].set_xlabel('Student')
axs[1, 0].set_ylabel('Banana')
axs[1, 0].set_title('Student vs. Banana Relationship')

axs[1, 1].plot(x,pow(y, 4))

axs[1, 1].set_xlabel('Student')
axs[1, 1].set_ylabel('Banana')
axs[1, 1].set_title('Student vs. Banana Relationship')

plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize = (12, 8))

axs[0, 0].plot(x,y)
axs[0, 0].set_xlabel('Student')
axs[0, 0].set_ylabel('Banana')
axs[0, 0].set_title('Student vs. Banana Relationship')

axs[0, 1].plot(x,pow(y, 2))

axs[0, 1].set_xlabel('Student')
axs[0, 1].set_ylabel('Banana')
axs[0, 1].set_title('Student vs. Banana Relationship')

axs[1, 0].plot(x,pow(y, 3))

axs[1, 0].set_xlabel('Student')
axs[1, 0].set_ylabel('Banana')
axs[1, 0].set_title('Student vs. Banana Relationship')

axs[1, 1].plot(x,pow(y, 4))

axs[1, 1].set_xlabel('Student')
axs[1, 1].set_ylabel('Banana')
axs[1, 1].set_title('Student vs. Banana Relationship')

#Tight Layout making it easier to stop the overlap in the graphs above
plt.tight_layout()
plt.show()
```

### How do we change the font size? (Title, axis labels, and tick labels)

```python
fig, ax = plt.subplots()
ax.plot(x,y)

ax.set_xlabel('Student', fontsize = 12)
ax.set_ylabel('Banana', fontsize = 12)
ax.set_title('Student vs. Banana Relationship', fontsize = 20)

plt.show()
```

```python
ax.tick_params?
```

```python
fig, ax = plt.subplots()
ax.plot(x,y)

ax.set_xlabel('Student', fontsize = 12)
ax.set_ylabel('Banana', fontsize = 12)
ax.set_title('Student vs. Banana Relationship', fontsize = 20)

ax.grid()
ax.tick_params(axis = 'both', labelsize = 15, grid_color ='r', grid_alpha = 0.4)

plt.show()
```

### How to remove ticks and tick labels?

```python
fig, ax = plt.subplots()
ax.plot(x,y)

ax.set_xticks([])
ax.set_yticks([])

plt.show()
```

### How to remove axis and grid lines?

```python
fig, ax = plt.subplots()
ax.plot(x,y)

ax.set_axis_off()

plt.show()
```

### How to rotate axis text?

```python
fig, ax = plt.subplots()
ax.plot(x,y)

ax.tick_params(axis = 'x', rotation = 45)

plt.show()
```

### Logarithmic Axis

```python
x = np.arange(1, 10, 0.1)
y = np.exp(x*30)

fig, ax = plt.subplots()
ax.plot(x,y)

ax.set_yscale('log')

plt.show()
```
